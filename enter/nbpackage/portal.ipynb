```python
%matplotlib notebook
import os
import ipywidgets as widgets
from IPython.display import display
from IPython.display import display_html
from wind_order.func_run import main_run
import IPython.core.display as di       # Example: di.display_html('<h3>%s:</h3>' % str, raw=True)

def run():
    # This line will hide code by default when the notebook is exported as HTML
    di.display_html('<script>jQuery(function() {if (jQuery("body.notebook_app").length == 0) { jQuery(".input_area").toggle(); jQuery(".prompt").toggle();}});</script>', raw=True)
    
    THIS_DIR = os.getcwd()
    REF_STD_FOLDER = os.path.abspath(os.path.join(THIS_DIR, "../enter/参考设计风参/标准设计风参/"))
    REF_CUS_FOLDER = os.path.abspath(os.path.join(THIS_DIR, "../enter/参考设计风参/定制化塔架设计风参/"))
    CUS_FOLDER = os.path.abspath(os.path.join(THIS_DIR, "../enter/项目场址风参/"))

    ref_std_list = [f for f in os.listdir(os.path.abspath(REF_STD_FOLDER)) if f.endswith('xlsx') or f.endswith('xls') ]
    ref_cus_list = [f for f in os.listdir(os.path.abspath(REF_CUS_FOLDER)) if f.endswith('xlsx') or f.endswith('xls') ]
    cus_list = [f for f in os.listdir(os.path.abspath(CUS_FOLDER)) if f.endswith('xlsx') or f.endswith('xls') ]
    
    ref_std_list.insert(0, '')
    ref_cus_list.insert(0, '')
    
    ref_wind_path = []
    ref_wind_path.extend(ref_std_list)
    ref_wind_path.extend(ref_cus_list)

    
    ref_std_files_select = widgets.SelectMultiple(
        options=ref_std_list,
        value=[ref_std_list[0]],
#         description='标准设计风参：',
        disabled=False
    )
    
    ref_cus_files_select = widgets.SelectMultiple(
        options=ref_cus_list,
        value=[ref_cus_list[0]],
#         description='定制化塔架设计风参：',
        disabled=False
    )

    cus_files_select = widgets.Dropdown(
        options=cus_list,
        value=cus_list[0],
#         description='项目场址风参：',
        width='50%',
        disabled=False,
    )

    run_btn = widgets.Button(description = "运行")
    def btn_click(sender):
        wind_path = os.path.join(CUS_FOLDER, cus_files_select.value)
        ref_std_wind_path = [os.path.join(REF_STD_FOLDER, d) for d in ref_std_files_select.value if d]
        ref_cus_wind_path = [os.path.join(REF_CUS_FOLDER, d) for d in ref_cus_files_select.value if d]
        
        ref_wind_path = []
        ref_wind_path.extend(ref_cus_wind_path)
        ref_wind_path.extend(ref_std_wind_path)
#         print(ref_wind_path, ref_cus_wind_path, ref_std_wind_path)
        
        if not os.path.isfile(wind_path):
            print('风参路径不正确！')
        else:
            main_run(THIS_DIR, wind_path, ref_wind_path)

    run_btn.on_click(btn_click)

    hbox1 = widgets.HBox([widgets.Label(value="项目场址风参      ："), cus_files_select], layout=widgets.Layout(alignt='stretch'))
    hbox2 = widgets.HBox([widgets.Label(value="标准设计风参      ："), ref_std_files_select], layout=widgets.Layout(align='stretch'))
    hbox3 = widgets.HBox([widgets.Label(value="定制化塔架设计风参："), ref_cus_files_select], layout=widgets.Layout(align='stretch'))
    LAYOUT = widgets.HBox([widgets.VBox([hbox1,hbox2,hbox3]), run_btn])

    display(LAYOUT)
```
